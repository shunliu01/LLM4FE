```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
event_data = pd.read_csv('../data/#2/event_data.csv')
item_data = pd.read_csv('../data/#2/item_data.csv')
sales_data = pd.read_csv('../data/#2/sales_data.csv')
site_data = pd.read_csv('../data/#2/site_data.csv')
```

```python
sales_data.sort_values('DS', inplace=True)
sales_data.reset_index(drop=True, inplace=True)
sales_data.head()
```

```python
import sweetviz as sv # Auto EDA

report = sv.analyze(sales_data)
report.show_html('../report/#2/sales_data_report.html')
```

```python
from autoviz.AutoViz_Class import AutoViz_Class
AVC = AutoViz_Class()

filename = "../data/#2/sales_data.csv" # Generating AutoViz Report #this is the default command when using a file for the dataset
sep = ","
dft = AVC.AutoViz(
    filename,
    sep=",",
    depVar="",
    dfte=None,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
    )
```
